```python
import pandas as pd
import numpy as np
```

```python
filepath='/content/Copy of customer_transactions_sample.csv'
data=pd.read_csv(filepath)
data
```

```python
data_uk=data[data['Country']=='United Kingdom']
data_uk
```

```python
data_uk.isnull().sum()
```

```python
data_uk=data[data['Quantity']>0]
data_uk
```

```python
#remove rows where customerID are NA
data_uk.dropna(subset=['Customer ID'],how='all',inplace=True)
print(data_uk.shape)
```

```python
data_uk['InvoiceDate']
#2009-12-01 07:45:00
#1        2009-12-01 07:45:00
#2        2009-12-01 07:45:00
#3        2009-12-01 07:45:00
#4        2009-12-01 07:45:00
```

```python
#converting into date format to work easily
data_uk['InvoiceDate'] = pd.to_datetime(data_uk['InvoiceDate'], errors='ignore')
data_uk['InvoiceDate']
```

```python
data_uk['InvoiceDate'].max()
```

```python
#now to calculate the recency we need to get the dates to know how many days before the transaction was made
import datetime as dt
last=dt.date(2010,12,9)
last
```

```python
#create a new column called date which contains the date of invoice only
data_uk['date'] = pd.DatetimeIndex(data_uk['InvoiceDate']).date
data_uk
```

```python
#we will cluster all over the year
data_uk_year=data_uk[data_uk['InvoiceDate']>= '01-04-2010  09:24:00']
```

```python
data_uk_year
```

```python
#now we have year data and ready to calculate the recency
recency_df=data_uk_year.groupby(by='Customer ID',as_index=False)['date'].max()
recency_df.columns = ['Customer ID','LastPurshaceDate']
recency_df
```

```python
recency_df['recency']=recency_df['LastPurshaceDate'].apply(lambda x: (last-x).days)
recency_df
```

```python
#now we will calculate the frequency
#invoice represent transaction of a customer
data_uk_year_copy=data_uk_year
data_uk_year_copy.drop_duplicates(subset=['Customer ID','Invoice'],keep="first",inplace=True)
data_uk_year_copy.head()
```

```python
#now we will calculate the frequency
#invoice represent transaction of a customer
freq_df=data_uk_year_copy.groupby(by=['Customer ID'],as_index=False)['Invoice'].count()
freq_df.columns=['Customer ID','Freq']
freq_df
```

```python
#now we will calculate the monetary
data_uk['Total price']=data_uk['Quantity']*data_uk['Price']
monetary_df=data_uk.groupby(by=['Customer ID'],as_index=False)['Total price'].sum()
monetary_df.columns=['Customer ID','Monetary']
monetary_df
```

```python
#now we have recency,monetary,frequency
#merging all to apply rfm model
rfm_df=recency_df.merge(freq_df,on='Customer ID')
rfm_df
```

```python
rfm_df=rfm_df.merge(monetary_df,on='Customer ID')
#set cutomer id as index
rfm_df.set_index('Customer ID',inplace=True)
rfm_df
```

```python
#now we will form rfm model and use quartiles(range 1-4) rather quintiles(range 1-5)
quantiles=rfm_df.quantile(q=[0.25,0.50,0.75])
quantiles
```

```python
quantiles.to_dict
```

```python
#for a good customer less recency and high frequency and high monetary
# x is value of recency/monetary/frequency and d is dictionary, p is value of customer
def Rscore(x,p,d):
  if x<=d[p][0.25]:
   return 4
  elif x <= d[p][0.50]:
   return 3
  elif x <=d[p][0.75]:
   return 2
  else:
   return 1
# x is value of recency/monetary/frequency and d is dictionary, p is value of customer
def Fscore(x,p,d):
  if x>=d[p][0.75]:
    return 4
  elif x>=d[p][0.50]:
    return 3
  elif x>=d[p][0.25]:
    return 2
  else:
    return 1
```

```python
rfm_segmentation =rfm_df
rfm_segmentation['Rscore']=rfm_segmentation['recency'].apply(Rscore, args=('recency',quantiles))
rfm_segmentation['Fscore']=rfm_segmentation['Monetary'].apply(Fscore,args=('Monetary',quantiles))
rfm_segmentation['Mscore']=rfm_segmentation['Freq'].apply(Fscore,args=('Freq',quantiles))
rfm_segmentation
```

```python
rfm_segmentation['RFM_score']=rfm_segmentation.Rscore.map(str) +rfm_segmentation.Fscore.map(str) + rfm_segmentation.Mscore.map(str)
rfm_segmentation
```

```python
#lets find  the our best customer means RFM_score =444
print("Best customer:",rfm_segmentation[rfm_segmentation['RFM_score']=='444'].shape[0])
```

```python
#now giving the result
print("Best customer:",rfm_segmentation[rfm_segmentation['RFM_score']=='444'].shape[0])
print("Big spenders customer:",rfm_segmentation[rfm_segmentation['Mscore']==4].shape[0])
print("Loyal customer:",rfm_segmentation[rfm_segmentation['Fscore']==4].shape[0])
print("Lost customer:",rfm_segmentation[rfm_segmentation['RFM_score']=='111'].shape[0])
```
